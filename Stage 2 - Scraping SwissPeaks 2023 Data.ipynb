### Sample: Liam Vines

```python
vines_initial_df = initial_runner_info_df[initial_runner_info_df['Name'] == 'Liam Vines']
vines_initial_df
```

```python
year = '2023'
vines_url = 'swip-2023-sp23-360-liam-vines'
vines_stations_list = []

# URL of the JSON file
url = f'https://livetrack.me/api/visitor/event_sessions/swip-2023/enroll_participants/{vines_url}/states'
print(url)
# Send a GET request to fetch the JSON data
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the JSON data
    data = response.json()

    # Optionally, save the data to a JSON file
    with open(f'Data/JSON/SwissPeaks_vines_{year}.json', 'w') as f:
        json.dump(data, f, indent=4)

    
    # Send a GET request to fetch the JSON data
    response = requests.get(url)
    
    if response.status_code == 200:
        # Parse the JSON data
        data = response.json()
        
        # Extract status if available
        status = data.get('message', 'No status message found')
        
        # Clean and format name
        name = vines_url.replace('swip-2023-sp23-360-', '')
        name = name.replace('swip-2023-sp23-360d-', '')
        name = name.replace('-', ' ').title()
        
        print("Name:", name)
        print("Status:", status)

        for data_from_scrape in  data['data']['states']:            
                bib = data_from_scrape['bib_number']
                station_id = data_from_scrape['station_id']
                timestamp = data_from_scrape['timestamp']

                if timestamp is not None:
                    timestamp = datetime.datetime.utcfromtimestamp(timestamp)
                else:
                    timestamp = None  # or set a default datetime, e.g. datetime.datetime.min

                stay_on_station = data_from_scrape['stay_on_station']
                penalty_in_seconds = data_from_scrape['penalty_in_seconds']        
                timestamp_from_start = data_from_scrape['timestamp_from_start']
                timestamp_from_previous = data_from_scrape['timestamp_from_previous']
                time_to_first = data_from_scrape['time_to_first']


                all_info = [bib, vines_url, status, 
                            station_id, timestamp, 
                            stay_on_station, penalty_in_seconds, 
                            timestamp_from_start,timestamp_from_previous, time_to_first]

        
                vines_stations_list.append(all_info)

    else:
        print(f"Failed to retrieve data. Status code: {response.status_code}")


            

vines_station_df =  pd.DataFrame(vines_stations_list, columns=['Bib','URL', 'Status', 
                                                'Station', 'Timestamp',
                                                'stay_on_station', 'penalty_in_seconds', 
                                                'timestamp_from_start','timestamp_from_previous',
                                                'time_to_first'])   
vines_station_df
        
```

```python
# Merge the datsets back together
vines_df = vines_station_df.merge(
     vines_initial_df, 
    on = 'URL',
    how='left'
)

vines_df['Race'] = np.where(
    vines_df['URL'].str.contains('swip-2023-sp23-360d-', na=False), 
    'Swiss Peaks 360 DUO',  'Swiss Peaks 360')

vines_df['Year'] = '2023'

vines_df = vines_df[[ "Name", "Age", 'Sex', 'Status', 
                'Station','Timestamp',
                  'Race', 'Year', 'Bib',   'URL', 
                'stay_on_station',
                'penalty_in_seconds', 'timestamp_from_start',
                'timestamp_from_previous', 'time_to_first']]

vines_df['Station'] = vines_df['Station'].str.replace('swip-2023-', '')
vines_df['Station'] = vines_df['Station'].str.replace('-', ' ')
vines_df['Station'] = vines_df['Station'].str.title()
vines_df['Station'] = vines_df['Station'].str.replace('Depart Ulrichen 1', 'Oberwald')
vines_df['Station'] = vines_df['Station'].str.replace('Depart Grande Dixence', 'Grande Dixence Bv OUT')
vines_df['Station'] = vines_df['Station'].str.replace('Arrivee Bouveret', 'Bouveret')


vines_df['Station'] = vines_df['Station'].str.replace('In', 'IN')
vines_df['Station'] = vines_df['Station'].str.replace('Out', 'OUT')

vines_df['Station'] = vines_df['Station'].str.replace('Bv', 'BV')


vines_df.head()
```

```python
lifebases = [
    'Oberwald',
    'Fiesch BV IN',
 'Fiesh BV OUT',
             
 'Eisten BV IN',
 'Eisten BV OUT',
             
 'Grimentz BV IN',
 'Grimentz BV OUT',
             
 'Grande Dixence BV IN',
 'Grande Dixence BV OUT',
             
 'Finhaut BV IN',
 'Finhaut BV OUT',
             
 'Morgins BV IN',
 'Morgins BV OUT',
'Bouveret']


lifebases_vines = vines_df[vines_df['Station'].isin(lifebases)]
# Set 'Station' as a categorical column with the defined order
lifebases_vines['Station'] = pd.Categorical(
        lifebases_vines['Station'],
        categories = lifebases,
        ordered=True
    )
    
lifebases_vines = lifebases_vines.sort_values(by = 'Station',  ascending = True)
lifebases_vines['Diff'] = lifebases_vines['Timestamp'].diff()

lifebases_vines['Duration'] = lifebases_vines['Timestamp'][lifebases_vines['Station'] == 'Bouveret'].values[0]- lifebases_vines['Timestamp'][lifebases_vines['Station'] == 'Oberwald'].values[0]

lifebases_vines[['Bib', 'Name', 'Age', 'Sex', 'Status','Duration',  'Station', 'Timestamp', 'Race', 'Year',
       
       'Diff']].reset_index(drop = True)
```

### Now extracting more runners

```python
runners_station_list = []
failed_to_fetch = []


headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0 Safari/537.36',
    'Accept': 'application/json',
}

for unique_runner_url in initial_runner_info_df['URL'].unique():
    print("Unique runner URL:", unique_runner_url)
#             'https://livetrack.me/api/visitor/event_sessions/swip-2023/enroll_participants/swip-2023-sp23-360-andrea-dell-oro/states
    # Construct the API URL
    url = f'https://livetrack.me/api/visitor/event_sessions/swip-2023/enroll_participants/{unique_runner_url}/states'
    print("API URL:", url)
    
    # Send a GET request to fetch the JSON data
    response = requests.get(url, headers=headers, timeout=10)
    
    if response.status_code == 200:
        # Parse the JSON data
        data = response.json()
        
        # Extract status if available
        status = data.get('message', 'No status message found')
        
        # Clean and format name
        name = unique_runner_url.replace('swip-2023-sp23-360-', '')
        name = name.replace('swip-2023-sp23-360d-', '')
        name = name.replace('-', ' ').title()
        
        print("Name:", name)
        print("Status:", status)

        for data_from_scrape in  data['data']['states']:            
                bib = data_from_scrape['bib_number']
                station_id = data_from_scrape['station_id']
                timestamp = data_from_scrape['timestamp']

                if timestamp is not None:
                    timestamp = datetime.datetime.utcfromtimestamp(timestamp)
                else:
                    timestamp = None  # or set a default datetime, e.g. datetime.datetime.min

                stay_on_station = data_from_scrape['stay_on_station']
                penalty_in_seconds = data_from_scrape['penalty_in_seconds']        
                timestamp_from_start = data_from_scrape['timestamp_from_start']
                timestamp_from_previous = data_from_scrape['timestamp_from_previous']
                time_to_first = data_from_scrape['time_to_first']


                all_info = [bib, unique_runner_url, status, 
                            station_id, timestamp, 
                            stay_on_station, penalty_in_seconds, 
                            timestamp_from_start,timestamp_from_previous, time_to_first]

                runners_station_list.append(all_info)

    else:
        print(f"Failed to fetch data for {unique_runner_url}. Status code: {response.status_code}")
        failed_to_fetch.append(unique_runner_url)
        
    
    #### So we dont get blocked!!!!
    delay_seconds = random.randint(10,60)
    print('Delay before calling the server again: ', delay_seconds)
    print('*'*20)
    time.sleep(delay_seconds)

```
