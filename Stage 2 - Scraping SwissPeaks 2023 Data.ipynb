## Extracting Data from JSON data

```python
# Prepare a list to hold the rows for the DataFrame
runner_info = []

year = '2023'
# Function to parse each entry
        # Load the JSON data from a file
with open(f'Data/JSON/SwissPeaks_{year}.json', 'r') as file:
    json_data = json.load(file)
    json_data_DATA = json_data['data']
    
    
    for data in json_data['data']:
#         print(data['enroll_participant_id'])
        
        unique_urls_for_each_runner = data["enroll_participant_id"]
        gender_category = data["enroll_participant_category"]
        gender = data["enroll_participant_gender"]
        name = unique_urls_for_each_runner.replace('swip-2023-sp23-360-', '' )
        name = name.replace('swip-2023-sp23-360d-', '' )
        name = name.replace('-', ' ' )
        name = name.title()
        print(name)
         
        all_info = [name, gender_category, gender, unique_urls_for_each_runner]         
#         # Append the row to the list
        runner_info.append(all_info)
    
```

```python
initial_runner_info_df =  pd.DataFrame(runner_info, columns=["Name", "Age", 'Sex', 'URL'])   
```

### Now extracting more runners

```python
initial_runner_info_df['URL'].unique()
```

```python
for unique_runner_url in initial_runner_info_df['URL'].unique()[:5]:
    print("Unique runner URL:", unique_runner_url)
```

```python
runners_station_list = []

for unique_runner_url in initial_runner_info_df['URL'].unique()[:5]:
    print("Unique runner URL:", unique_runner_url)
    
    # Construct the API URL
    url = f'https://livetrack.me/api/visitor/event_sessions/swip-2023/enroll_participants/{unique_runner_url}/states'
    print("API URL:", url)
    
    # Send a GET request to fetch the JSON data
    response = requests.get(url)
    
    if response.status_code == 200:
        # Parse the JSON data
        data = response.json()
        
        # Extract status if available
        status = data.get('message', 'No status message found')
        
        # Clean and format name
        name = unique_runner_url.replace('swip-2023-sp23-360-', '')
        name = name.replace('swip-2023-sp23-360d-', '')
        name = name.replace('-', ' ').title()
        
        print("Name:", name)
        print("Status:", status)

        for data_from_scrape in  data['data']['states']:            
                bib = data_from_scrape['bib_number']
                station_id = data_from_scrape['station_id']
                timestamp = data_from_scrape['timestamp']

                if timestamp is not None:
                    timestamp = datetime.datetime.utcfromtimestamp(timestamp)
                else:
                    timestamp = None  # or set a default datetime, e.g. datetime.datetime.min

                stay_on_station = data_from_scrape['stay_on_station']
                penalty_in_seconds = data_from_scrape['penalty_in_seconds']        
                timestamp_from_start = data_from_scrape['timestamp_from_start']
                timestamp_from_previous = data_from_scrape['timestamp_from_previous']
                time_to_first = data_from_scrape['time_to_first']


                all_info = [bib, unique_runner_url, status, 
                            station_id, timestamp, 
                            stay_on_station, penalty_in_seconds, 
                            timestamp_from_start,timestamp_from_previous, time_to_first]

                runners_station_list.append(all_info)

    else:
        print(f"Failed to fetch data for {unique_runner_url}. Status code: {response.status_code}")
        
    
    #### So we dont get blocked!!!!
    delay_seconds = random.randint(10,40)
    print(delay_seconds)
    time.sleep(delay_seconds)
    print('*'*20)
```

### Rearranging and saving

```python
runner_info_df =  pd.DataFrame(runners_station_list, columns=[ 'Bib','URL', 'Status', 
                                                'Station', 'Timestamp',
                                                'stay_on_station', 'penalty_in_seconds', 
                                                'timestamp_from_start','timestamp_from_previous',
                                                'time_to_first'])   

```

```python
# Merge the datsets back together
SP_2023  = initial_runner_info_df.merge(
    runner_info_df, 
    on = 'URL',
    how='left'
)
```
