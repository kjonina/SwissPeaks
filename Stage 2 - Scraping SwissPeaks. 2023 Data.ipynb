# SwissPeaks Data

In this code, Swiss Peaks website is scraped. The raw data is then saved to a folder with race and year. No cleaning is done.


```python
import requests
import json
import html
import pandas as pd
import time
import re
import unicodedata
import datetime
import random
```

## Scraping TORX Data from the website

```python
# https://livetrack.me/api/visitor/event_sessions/swip-2023/races/swip-2023-sp23-360/ranking
```

```python
year = '2023'

delay_seconds = 15

# URL of the JSON file
url = f'https://livetrack.me/api/visitor/event_sessions/swip-{year}/races/swip-{year}-sp23-360/ranking'
print(url)

# Send a GET request to fetch the JSON data
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the JSON data
    data = response.json()

    # Optionally, save the data to a JSON file
    with open(f'Data/JSON/SwissPeaks_{year}.json', 'w') as f:
        json.dump(data, f, indent=4)

    print(f"Data saved to 'SwissPeaks_{year}.json'")
else:
    print(f"Failed to retrieve data. Status code: {response.status_code}")

time.sleep(delay_seconds)
```

## Extracting Data from JSON data

```python
# Prepare a list to hold the rows for the DataFrame
runner_info = []

# Function to parse each entry
        # Load the JSON data from a file
with open(f'Data/JSON/SwissPeaks_{year}.json', 'r') as file:
    json_data = json.load(file)
    json_data_DATA = json_data['data']
    
    
    for data in json_data['data']:
#         print(data['enroll_participant_id'])
        
        unique_urls_for_each_runner = data["enroll_participant_id"]
        gender_category = data["enroll_participant_category"]
        gender = data["enroll_participant_gender"]
        name = unique_urls_for_each_runner.replace('swip-2023-sp23-360-', '' )
        name = name.replace('swip-2023-sp23-360d-', '' )
        name = name.replace('-', ' ' )
        name = name.title()
        print(name)
         
        all_info = [name, gender_category, gender, unique_urls_for_each_runner]         
#         # Append the row to the list
        runner_info.append(all_info)
    
```

```python
initial_runner_info_df =  pd.DataFrame(runner_info, columns=["Name", "Age", 'Sex', 'URL'])   
```
